```python
#Import Libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Read the data set
df = pd.read_csv("Housing.csv")
```

1.  SCATTER PLOT

```python

plt.ylim(0, 300000)
plt.scatter(df.type, df.price, color = 'hotpink')

plt.xlabel("Type",  size=14)
plt.ylabel("Price" ,  size=14)
plt.title("Scatter Plot of Type Vs Price ",  size=16)
```

```python
plt.xlim(0, 200000)
plt.ylim(0, 5)

plt.scatter(df.price,df.beds, color = 'lightgreen')

plt.xlabel("Price",size=14)
plt.ylabel("No of Beds",size=14)
plt.title("Scatter Plot of Price Vs No of Bed Rooms",size=16)
```

2.  PIE CHART

```python

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'apartment', 'condo', 'house', 'duplex', 'townhouse'
sizes = [22.073267,21.430631,20.780202,18.182645, 17.533255 ]
explode = (0.15,0, 0, 0, 0)  # only "explode" the 2nd slice 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode,  autopct='%1.1f%%',
        shadow=True, startangle=90,textprops={'fontsize': 14} )
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

legend = plt.legend(labels ,title = "Types:" ,bbox_to_anchor=(1,1))

legend.get_title().set_fontsize('16') #legend 'Title' fontsize
plt.setp(plt.gca().get_legend().get_texts(), fontsize='14') #legend 'list' fontsize

plt.title("Types Pie Chart" , size=16)

```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ["No", "Yes"]

sizes = [95.187245,4.812755]
explode = (0.15,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode,autopct='%1.1f%%',
        shadow=True,labels=labels,startangle=90,textprops={'fontsize': 14} )
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("Comes with Furnitures Pie Chart" , size=16)
```

3. DONUT CHART

```python
import matplotlib.pyplot as plt
# Pie chart
labels = ['w/d in unit', 'w/d hookups', 'laundry on site ', 'laundry in bldg', 'no laundry on site']
sizes = [43.073237,24.699380,19.242624,11.800256, 1.184503]
#colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#B7C3F3']
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors,autopct='%1.2f%%', startangle=90,textprops={'fontsize': 14})
#draw circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()

legend = plt.legend(labels ,title = "Laundry Options :" ,bbox_to_anchor=(1,1))

legend.get_title().set_fontsize('16') #legend 'Title' fontsize
plt.setp(plt.gca().get_legend().get_texts(), fontsize='14') #legend 'list' fontsize

plt.title("Laundry Options Donut Chart" , size=16)
plt.show()
```

3. BAR PLOT

```python

fig = plt.figure(figsize=(20, 8), dpi=80)
ax = fig.add_axes([0,0,1,1])

lables = ['off-street parking', 'attached garage', 'carport', 'detached garage', 'street parking','no parking','valet parking']
values = [128502,40591,38955,16940,15951,3188,163]

plt.xlabel('Parking Options', fontsize=30)
plt.ylabel('Count', fontsize=30)

plt.rcParams['font.size'] = '20'

ax.bar(lables,values, color = '#A061B4')
plt.show()

```
